```python
# importing all the necessary libraries
import selenium 
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
PATH = "C:\Program Files (x86)\chromedriver.exe"
from selenium import webdriver
driver = webdriver.Chrome(PATH)
driver.get("https://www.deccanherald.com/sitemap/detail/year/2021-3")
```

```python
# scraping the date url 
date_url = []
for i in driver.find_elements_by_xpath('//li[@class="taxonomy-term children"]/a'):
    date_url.append(i.get_attribute('href'))
```

```python
len(date_url)
```

```python
# scraping the heading url 
Heading_url = []
for j in date_url:
    driver.get(j)
    try:
        for i in driver.find_elements_by_xpath('//div[@class="group"]/ul/li/a'):
            Heading_url.append(i.get_attribute('href'))
    except NoSuchElementException:
           Heading_url.append('--')          
```

```python
len(Heading_url)
```

```python
date = []
author = []
headline = []
vertical = []
description = []
for k in Heading_url:
    driver.get(k)
    
    # Scraping Date of News
    
    try:
        for i in driver.find_elements_by_xpath('//li[@class="sanspro-semib text-uppercase crud-items__lists"][1]'):
            date.append(i.text)
    except NoSuchElementException:
            date.append('--')
            
    # Scraping Author of News
    
    try:
        for i in driver.find_elements_by_xpath('//div[@class="article-author__time"]/a'):
            author.append(i.text)
    except NoSuchElementException:
            author.append('--')      
            
    # Scraping Verticals of News
    
    try:
        for i in driver.find_elements_by_xpath('//div[@class="item-list"]/ul/li[2]/a'):
            vertical.append(i.text)
    except NoSuchElementException:
            vertical.append('--')      
            
            
    # Scraping Headline of News
    
    try:
        for i in driver.find_elements_by_xpath('//div[@class="strap-heading-wrapper"]/h1'):
            headline.append(i.text)
    except NoSuchElementException:
            headline.append('--')    
            
            
    # Scraping Description of News
    
    try:
        for i in driver.find_elements_by_xpath('//div[@class="field field-name-body field-type-text-with-summary field-label-hidden"]'):
            description.append(i.text)
    except NoSuchElementException:
            description.append('--')             
            
```

```python
print(len(date),len(author),len(headline),len(vertical),len(description))
```

```python
import pandas as pd 
df = pd.DataFrame({
    'Date':date[:10368],
    'Author':author[:10368],
    'Headline':headline[:10368],
    'Vertical':vertical[:10368],
    'Description':description[:10368]
    
})
df
```

```python
df.replace({'Author':[',']},'',regex = True,inplace = True)
```

```python
# converting df to xml file 
xml_data = ['<root>']
for column in df.columns:
    xml_data.append('<{}>'.format(column))  # Opening element tag
    for field in df.index:
        # writing sub-elements
        xml_data.append('<{0}>{1}</{0}>'.format(field, df[column][field]))
    xml_data.append('</{}>'.format(column))  # Closing element tag
xml_data.append('</root>')

with open('C:\\Users\\user\\OneDrive\\Documents\\Deccan_Herald.xml', 'w',encoding="utf-8") as f:  # Writing in XML file
    for line in xml_data:
        f.write(line)
```
